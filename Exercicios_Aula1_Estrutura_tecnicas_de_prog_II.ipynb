```python
import pandas as pd
import matplotlib.pyplot as plt
df_CO2 = pd.read_csv('/content/co2_emissions_kt_by_country.csv')
df_CO2
```

```python
df_brazil_CO2 = df_CO2[df_CO2['country_name'] == 'Brazil']
df_brazil_CO2
```

```python
df_brazil_grouped = df_brazil_CO2.groupby('year')['value'].mean()

# Cria um gráfico de linha com os valores médios de cada ano
plt.plot(df_brazil_grouped.index, df_brazil_grouped.values)

plt.title('Emissões de CO2 no Brasil por ano')
plt.xlabel('Ano')
plt.ylabel('Emissões de CO2 (em kt)')
plt.show()
```

```python
df_media_mundo_CO2 = df_CO2.groupby('year')['value'].mean()

plt.plot(df_media_mundo_CO2.index, df_media_mundo_CO2.values)

plt.title('Emissões de CO2 Mundial por ano')
plt.xlabel('Ano')
plt.ylabel('Emissões de CO2 (em kt)')
plt.show()
```

```python
fig, ax = plt.subplots()

df_brazil_grouped = df_brazil_CO2.groupby('year')['value'].mean()
ax.plot(df_brazil_grouped.index, df_brazil_grouped.values, label='Brasil')

df_media_mundo_CO2 = df_CO2.groupby('year')['value'].mean()
ax.plot(df_media_mundo_CO2.index, df_media_mundo_CO2.values, label='Mundo')

ax.set_title('Emissões de CO2 por ano')
ax.set_xlabel('Ano')
ax.set_ylabel('Emissões de CO2 (em kt)')
ax.legend()

# Exibe o gráfico
plt.show()
```

```python
df_2018_2019 = df_CO2[df_CO2['year'].isin([2018, 2019])]

df_total_emissions = df_2018_2019.groupby(['country_name', 'year'])['value'].sum().unstack()
df_total_emissions
```

```python
df_percent = ((df_total_emissions[2019] - df_total_emissions[2018]) / df_total_emissions[2018]) * 100
df_percent
```

```python
df_percemt_ordenado = df_percent.sort_values(ascending=False)
df_percemt_ordenado.head(10)
```

```python
plt.barh(df_percemt_ordenado.head(10), df_percemt_ordenado.head(10))
plt.title('Os 10 países que mais aumentaram a emissão de CO2 entre 2018 e 2019')
plt.xlabel('Aumento percentual de emissão de CO2')
plt.ylabel('País')
plt.show()
```
