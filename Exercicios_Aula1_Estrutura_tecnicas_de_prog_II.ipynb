Exercício 1. Reproduza o gráfico de barras abaixo, em que cada barra representa um dos 10 países com mais casos confirmados de COVID no dataset, e a "quebra" em cada cor indica a predominância de casos confirmados em cada mês.

image.png

```python
import pandas as pd
df_covid = pd.read_csv("https://raw.githubusercontent.com/MatheusRobles/Santander_Coders_Tecnicas_de_Prog_II/main/Ex_Aula_1/Covid_19_Countrywise_timeseries.csv")
```

```python
df_covid.head()
```

```python
df_covid['Mes'] = pd.to_datetime(df_covid['ObservationDate']).dt.strftime('%Y-%m')
df_covid
```

```python
matriz = df_covid.groupby(["country", "Mes"])['New Confirmed'].sum().unstack().fillna(0)
```

```python
matriz = matriz.sort_values('2020-03', ascending=False)[:10]
matriz.plot(kind = 'barh', stacked=True)
```

EX 2. Reproduza o gráfico de linhas abaixo, que representa a série temporal da evolução de emissões de CO2 no Brasil.

image.png

```python
import pandas as pd
df_CO2 = pd.read_csv('https://raw.githubusercontent.com/MatheusRobles/Santander_Coders_Tecnicas_de_Prog_II/main/Ex_Aula_1/co2_emissions_kt_by_country.csv')
df_CO2
```

```python
df_brazil_CO2 = df_CO2[df_CO2['country_name'] == 'Brazil']
df_brazil_CO2
```

```python
df_brazil_grouped = df_brazil_CO2.groupby('year')['value'].mean()
df_brazil_grouped.plot(x = df_brazil_grouped.index, y = df_brazil_grouped.values)
```

EX 4. Os 10 países que mais aumentaram, percentualmente, a emissão de CO2 entre os anos de 2018 e 2019, na forma de um gráfico de barras horizontal.

image.png

```python
df_2018_2019 = df_CO2[df_CO2['year'].isin([2018, 2019])]

df_total_emissions = df_2018_2019.groupby(['country_name', 'year'])['value'].sum().unstack()
df_total_emissions
```

```python
df_total_emissions['percent']  = ((df_total_emissions[2019] - df_total_emissions[2018]) / df_total_emissions[2018]) * 100
df_total_emissions
```

```python
df_percent = df_total_emissions.sort_values(by='percent', ascending=False)
df_percent
```

```python
df_percent = df_percent.head(10)
df_percent = df_percent.sort_values(by='percent', ascending=True)
df_percent
```

```python
df_percent.head(10).percent.plot(kind = 'barh', title = "Top 10 países com maiores aumentos percentuais de emissão entre 2018 e 2019", )
```
